```python
# evaluate datatype 
col_list = ['attracts', 'flower', 'fruit', 'leaf', 'water', 'height', 'width', 'special_feature', 'soil_conditions', 'texture', 'temperature', 'emergence_days','maintenance', 'native_region', 'propagation', 'bloom_time', 'sun','problem_solvers', 'type', 'zones', 'color']
for col in col_list:
    df[col] = df[col].apply(eval)
df
```

```python
# total unique type of plants 3244
len(df.common.unique())  
```

```python
# expand each feature in list to a new row for further feature engineering
explode_cols = df.columns.to_list()
explode_cols.remove('common')  # common is plant name

for col in explode_cols:
    df = df.explode(col)
df.drop_duplicates(inplace=True)  # remove dup
df.head()
```

```python
# bucketize temperature base on criteria - Cold: 55 degrees and below ; Cool: 56 to 65 degrees ; Mild: 66 to 75 degrees ; Warm: 75 to 85 degrees ; Hot: Above 85 degrees.

df['temperature'] = df['temperature'].astype('int')

def bucketize(x):
    if x in range(55):
        x = 'low'
    elif x in range(56, 66):
        x = 'cool'
    elif x in range(66, 75):
        x = 'mild'
    elif x in range(75, 85):
        x = 'warm'
    elif x in range(85, 140):
        x = 'hot'
    else: 
        x = 'none'
    return x

df['temp_bucket'] = df.apply(lambda x: bucketize(x['temperature']),axis=1)
```

```python
df.drop(columns='temperature', inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
features = df.columns.to_list()
drop_list = ['common', 'description','family','native_region','scientific_name','genus', 'color']
for col in drop_list:
    features.remove(col)
    
len(features)

# no clear distribution found in 'family','native_region','scientific_name','genus', 'color' 
# these feature may require further cleanup
```

```python
# look at feature distributions 
plt.figure(figsize=(20, 20))
plt.subplots_adjust(hspace=1)
plt.suptitle("Feature Distribution", fontsize=18, y=0.95)

n_features = len(features)
ns_cols = 3
ns_rows = int(n_features/ns_cols)+1
# fig, axes = plt.subplots(nrows=ns_rows,ncols=ns_cols, figsize=(30,30))

for i, col in enumerate(features):
#     ax = axes[int(i/ns_cols),i%ns_cols]
    ax = plt.subplot(ns_rows, ns_cols, i+1)
    distribution = pd.DataFrame(df.groupby(col)['common'].count().sort_values())
    distribution.plot.bar(y='common', ylabel = 'counts', title=f'Num of Plants Per {col}', ax=ax)
```
