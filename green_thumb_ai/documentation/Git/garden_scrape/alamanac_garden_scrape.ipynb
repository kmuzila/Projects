```python
base_url = 'https://www.almanac.com'
url = 'https://www.almanac.com/gardening/growing-guides'
req = urllib.request.Request(url,headers={'User-Agent' : 'Mozilla/5.0'})
webpage = urllib.request.urlopen(req).read()
soup  = BeautifulSoup(webpage, 'html.parser')
```

```python
plant_links = []
a_links = soup.find_all('a', hreflang=True)
for a in a_links:
    plant_links.append(base_url+a.get('href'))
```

```python
req = urllib.request.Request(plant_links[0],headers={'User-Agent' : 'Mozilla/5.0'})
webpage = urllib.request.urlopen(req).read()
soup  = BeautifulSoup(webpage, 'html.parser')
soup
```
