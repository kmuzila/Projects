```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib
import requests 
import re
```

```python
url = requests.get('https://www.bhg.com/gardening/plant-dictionary/')
soup  = BeautifulSoup(url.text, 'html.parser') 
letter_links = soup.find_all('a', class_="alphabetical-nav__list-item-link button--outlined-little type--rabbit-bold")
soup_links  = soup.find_all('a', class_="link-list__link type--dog-bold type--dog-link")
```

```python
#  <a class="alphabetical-nav__list-item-link button--outlined-little type--rabbit-bold" href="#alphabetical-list-a">A</a>,
#   <a class="link-list__link type--dog-bold type--dog-link" href="https://www.bhg.com/gardening/plant-dictionary/annual/african-daisy/" rel="nocaes"> African Daisy
  
```

```python
plant_links = []
for link in soup_links:
    plant_links.append(link.get('href'))
```

```python
df = pd.DataFrame(columns=['title', 'genus', 'common', 'type', 'light', 'height', 'width', 'flower_color', 'foliage_color', 'season_feature', 'special_feature', 'zones', 'propagation', 'problem_solvers'])
```

```python
plant_raw = []
for link in plant_links:
    url = requests.get(link)
    soup  = BeautifulSoup(url.text, 'html.parser') 
    title = soup.find("title").text
    if soup.find("td", text="Description") != None:
        description = soup.find("td", text="Description").find_next_sibling("td").text
    else:
        description = None
    if soup.find("td", text="Genus Name") != None: 
        genus = soup.find("td", text="Genus Name").find_next_sibling("td").text
    else:
        genus = None
    if soup.find("td", text="Common Name") != None:
        common = soup.find("td", text="Common Name").find_next_sibling("td").text
    else:
        common = None
    if soup.find("td", text="Plant Type") != None:
        type = soup.find("td", text="Plant Type").find_next_sibling("td").text
    else:
        type = None
    if soup.find("td", text="Light") != None:
        light = soup.find("td", text="Light").find_next_sibling("td").text
    else:
        light = None
    if soup.find("td", text="Height") != None:
        height = soup.find("td", text="Height").find_next_sibling("td").text
    else:
        height = None
    if soup.find("td", text="Width") != None:
        width = soup.find("td", text="Width").find_next_sibling("td").text
    else:
        width = None
    if soup.find("td", text="Flower Color") != None:
        flower_color = soup.find("td", text="Flower Color").find_next_sibling("td").text
    else:
        flower_color = None
    if soup.find("td", text="Flower Color") != None:
        foliage_color = soup.find("td", text="Flower Color").find_next_sibling("td").text
    else:
        foliage_color = None
    if soup.find("td", text="Season Features") != None:
        season_feature = soup.find("td", text="Season Features").find_next_sibling("td").text
    else:
        season_feature = None
    if soup.find("td", text="Special Features") != None:
        special_feature = soup.find("td", text="Special Features").find_next_sibling("td").text
    else:
        special_feature = None
    if soup.find("td", text="Zones") != None:
        zones = soup.find("td", text="Zones").find_next_sibling("td").text
    else:
        zones = None
    if soup.find("td", text="Propagation") != None:
        propagation = soup.find("td", text="Propagation").find_next_sibling("td").text
    else:
        propagation = None
    if soup.find("td", text="Problem Solvers") != None:
        problem_solvers = soup.find("td", text="Problem Solvers").find_next_sibling("td").text
    else:
        problem_solvers = None
    df = df.append({'title' : title, 'genus' : genus, 'common' : common, 'type' : type, 'light' : light, 'height' : height, 'width' : width, 'flower_color' : flower_color, 'foliage_color' : foliage_color, 'season_feature' : season_feature, 'special_feature' : special_feature, 'zones' : zones, 'propagation' : propagation, 'problem_solvers' : problem_solvers}, ignore_index=True)
```
