```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib
import requests 
import re
```

```python
# https://github.com/khivasarap/GardenPlanner/blob/main/garden.ipynb
```

```python
url = requests.get('https://sympathink.com/vegetables/planting-calendar/')
soup  = BeautifulSoup(url.text, 'lxml') 
```

```python
headers = []
for li_tag in soup.find_all("ul",{"class":"da-filter-top"}):
    for span_tag in li_tag.find_all('li'):
        headers.append(span_tag.find('span', {'class':'sort'}).text.strip())
```

```python
df = pd.DataFrame(columns = headers)
df['Sowing Depth'] = ''
df['Planting Distance'] =''
df['Soil Type']=''
df['Family']=''
df['Group'] =''
df['Diseases']=''
df['Pests']=''
df['Common Name']=''
df['Moisture']=''
df['Scientific Name']=''
df = df.rename(columns={"A/Z": "Name"})
```

```python
def extract_elements(soup, rowIndex):
    for ultag in soup.find_all("ul",{"class":"vg-results vg-rows search-filter vg-calendar"}):
        for litag in ultag.find_all("li",{"class":"my-row vg-block"}):
            df.loc[rowIndex, 'Common Name'] = litag['data-c-name']
            df.loc[rowIndex, 'Common Name'] = litag['data-c-name']
            df.loc[rowIndex, 'Type'] = litag['data-cat']
            df.loc[rowIndex, 'Height'] = litag['data-hei']
            df.loc[rowIndex, 'Sowing Depth'] = litag['data-dep']
            df.loc[rowIndex, 'Planting Distance'] = litag['data-dis']
            df.loc[rowIndex, 'Diseases'] = litag['data-dse']
            df.loc[rowIndex, 'Pests'] = litag['data-pes']
            df.loc[rowIndex, 'Family'] = litag['data-fam']
            df.loc[rowIndex, 'Group'] = litag['data-gr']
            df.loc[rowIndex, 'Sun'] = litag['data-sush']
            df.loc[rowIndex, 'Soil Type'] = litag['data-s-ty']
            df.loc[rowIndex, 'Soil PH'] = litag['data-ph']
            df.loc[rowIndex, 'Scientific Name'] = litag['data-s-name']
            df.loc[rowIndex, 'Name'] = litag['data-e-name']
            df.loc[rowIndex, 'Moisture'] = litag['data-moi']
            df.loc[rowIndex, 'When to plant'] = litag['data-a-7']
            df.loc[rowIndex, 'When to harvest'] = litag['data-b-7']
            zone = litag.find("div",{"class":"vg-sub-7 ds-usda"}).text.strip()
            df.loc[rowIndex, 'Zone'] = zone
            rowIndex = rowIndex + 1
```

```python
rowIndex = 0
extract_elements(soup,rowIndex)
```
