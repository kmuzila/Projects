```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
mbg_df = pd.read_csv(r"C:\Users\Karl\Downloads\garden_scrape\mbg_df.csv")
veggies_df = pd.read_csv(r"C:\Users\Karl\Downloads\garden_scrape\veggies_df.csv")
bhg_df = pd.read_csv(r"C:\Users\Karl\Downloads\garden_scrape\bhg_df.csv")
```

```python
# season_feature needs to be parsed
# special_feature  needs to be parsed
# parse and fill bloom time
bhg_df.drop('Unnamed: 0', axis=1, inplace=True)
bhg_df.rename(str.lower, axis='columns', inplace=True)

bhg_df['width'].fillna('0 to 0 feet', inplace=True)
bhg_df['width'] = bhg_df['width'].str.replace('null','0')
bhg_df['width_ind'] = bhg_df['width'].str.split().str[0::2]
bhg_df['width_unit'] = bhg_df['width'].str.split().str[-1]
bhg_df['width_ind'] = bhg_df['width_ind'].apply(lambda x: list(map(int, x)))
bhg_df['width'] = bhg_df['width_ind'].apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))
bhg_df['width'] = bhg_df.apply(lambda x : np.divide(x.width, 12) if x.width_unit == 'inches' else x.width, axis=1)
bhg_df['width_unit'] = 'feet'

bhg_df['height'].fillna('0 to 0 feet', inplace=True)
bhg_df['height'] = bhg_df['height'].str.replace('null','0')
bhg_df['height_ind'] = bhg_df['height'].str.split().str[0::2]
bhg_df['height_unit'] = bhg_df['height'].str.split().str[-1]
bhg_df['height_ind'] = bhg_df['height_ind'].apply(lambda x: list(map(int, x)))
bhg_df['height'] = bhg_df['height_ind'].apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))
bhg_df['height'] = bhg_df.apply(lambda x : np.divide(x.height, 12) if x.height_unit == 'inches' else x.height, axis=1)
bhg_df['height_unit'] = 'feet'


bhg_df['light'].fillna('none', inplace=True)
bhg_df['light'] = bhg_df['light'].str.strip().str.lower().str.replace(', ', ' to ').str.replace('to sun', 'to full sun').str.replace('to shade', 'to full shade').apply(lambda x : ('full ' + x) if (x.startswith('sun') or x.startswith('shade')) else x)
bhg_df['light'] = bhg_df['light'].str.split(' to ')

bhg_df['zones'].fillna('0', inplace=True)
bhg_df['zones'] = bhg_df['zones'].str.strip().str.lower().str.split(', ')

bhg_df['flower_color'].fillna('none', inplace=True)
bhg_df['flower_color'] = bhg_df['flower_color'].str.strip().str.lower().str.split(', ')

bhg_df['season_feature'].fillna('none', inplace=True)
bhg_df['season_feature'] = bhg_df['season_feature'].str.strip().str.lower().str.split(', ')

bhg_df['special_feature'].fillna('none', inplace=True)
bhg_df['special_feature'] = bhg_df['special_feature'].str.strip().str.lower().str.split(', ')

bhg_df['propagation'].fillna('none', inplace=True)
bhg_df['propagation'] = bhg_df['propagation'].str.strip().str.lower().str.split(', ')

bhg_df['problem_solvers'].fillna('none', inplace=True)
bhg_df['problem_solvers'] = bhg_df['problem_solvers'].str.strip().str.lower().str.split(', ')

bhg_df['type'].fillna('none', inplace=True)
bhg_df['type'] = bhg_df['type'].str.strip().str.lower().str.split(', ')

bhg_df['common'] = bhg_df['common'].str.strip().str.lower().fillna('none')
bhg_df['genus'] = bhg_df['genus'].str.strip().str.lower().fillna('none')

bhg_df.rename(columns={"light": "sun", "season_feature" : "bloom_time", "flower_color" : "color"}, inplace=True)
bhg_df.drop(columns=['title','foliage_color','width_ind','width_unit','height_ind','height_unit'], inplace=True)

bhg_df = bhg_df.convert_dtypes()
```

```python
mbg_df.drop('Unnamed: 0', axis=1, inplace=True)
mbg_df.rename(str.lower, axis='columns', inplace=True)

mbg_df['zone'].fillna('0 to 0', inplace=True)
mbg_df['zone'] = mbg_df['zone'].apply(lambda x : '0 to '.join(x) if len(x) == 2 else x)
mbg_df['zone_ind'] = mbg_df['zone'].str.split().str[0::2]
mbg_df['zone_ind'] = mbg_df['zone_ind'].apply(lambda x: list(map(int, x)))
mbg_df['zone']  = mbg_df['zone_ind'].apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))

mbg_df['flower'].fillna('none', inplace=True)
mbg_df['flower'] = mbg_df['flower'].str.strip().str.lower().str.split(', ')

mbg_df['tolerate'].fillna('none', inplace=True)
mbg_df['tolerate'] = mbg_df['tolerate'].str.strip().str.lower().str.split(', ')

mbg_df['sun'].fillna('none', inplace=True)
mbg_df['sun'] = mbg_df['sun'].str.strip().str.lower().str.split(' to ')

mbg_df['fruit'].fillna('none', inplace=True)
mbg_df['fruit'] = mbg_df['fruit'].str.strip().str.lower().str.split(', ')

mbg_df['attracts'].fillna('none', inplace=True)
mbg_df['attracts'] = mbg_df['attracts'].str.strip().str.lower().str.split(', ')

mbg_df['leaf'].fillna('none', inplace=True)
mbg_df['leaf'] = mbg_df['leaf'].str.strip().str.lower().str.split(', ')

mbg_df['bloom description'].fillna('none', inplace=True)
mbg_df['bloom description'] = mbg_df['bloom description'].str.replace('Non', 'none').str.replace(' or ', ' to ').str.replace(' and ', ' to ').str.strip().str.lower().str.split(' to ')

mbg_df['suggested use'].fillna('none', inplace=True)
mbg_df['suggested use'] = mbg_df['suggested use'].str.strip().str.lower().str.split(', ')

mbg_df['native range'].fillna('none', inplace=True)
mbg_df['native range'] = mbg_df['native range'].str.strip().str.lower().str.split(r'(, )|( to )')

mbg_df['water'].fillna('none', inplace=True)
mbg_df['water'] = mbg_df['water'].str.strip().str.lower().str.split(' to ')

mbg_df['bloom time'].fillna('none', inplace=True)	
mbg_df['bloom time'] = mbg_df['bloom time'].str.strip().str.lower().str.split(' to ')	

mbg_df['common name'] = mbg_df['common name'].str.strip().str.lower().fillna('none')
mbg_df['formal name'] = mbg_df['formal name'].str.strip().str.lower().fillna('none')
mbg_df['family'] = mbg_df['family'].str.strip().str.lower().fillna('none')
mbg_df['type'] = mbg_df['type'].str.strip().str.lower().fillna('none').apply(lambda x: [x])
mbg_df['maintenance'] = mbg_df['maintenance'].str.strip().str.lower().fillna('none').str.split()

mbg_df['height'] = mbg_df['height'].fillna(0).apply(lambda x: [x])
mbg_df['spread'] = mbg_df['spread'].fillna(0).apply(lambda x: [x])

mbg_df.rename(columns={"formal name": "genus","common name": "common", "spread" : "width", "bloom description" : "color", "zone" : "zones",'native range' : 'native_region', 'tolerate' : 'problem_solvers', "bloom time" : "bloom_time", "suggested use" : "special_feature"}, inplace=True)
mbg_df.drop(columns=['garden uses','invasive','noteworthy characteristics','other','problems','culture','zone_ind'], inplace=True)

mbg_df = mbg_df.convert_dtypes()
```

```python
veggies_df.drop('Unnamed: 0', axis=1, inplace=True)
veggies_df.rename(str.lower, axis='columns', inplace=True)

veggies_df['emergence days'].fillna('0 to 0', inplace=True)
veggies_df['emergence days'] = veggies_df['emergence days'].str.strip().str.lower().str.split(' - ').str[0]
veggies_df['emergence days'] = veggies_df['emergence days'].str.split(' to ')
veggies_df['emergence days'] = veggies_df['emergence days'].apply(lambda x: list(map(int, x)))
veggies_df['emergence days'] = veggies_df['emergence days'].apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))

veggies_df['hardiness'].fillna('0 to 0', inplace=True)
veggies_df['hardiness'] = veggies_df['hardiness'].apply(
    lambda x: '0 to '.join(x) if len(x) == 2 else x)
veggies_df['hardiness_ind'] = veggies_df['hardiness'].str.split().str[0::2]
veggies_df['hardiness_ind'] = veggies_df['hardiness_ind'].apply(
    lambda x: list(map(int, x)))
veggies_df['hardiness'] = veggies_df['hardiness_ind'].apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))

veggies_df['sunlight'].fillna('none', inplace=True)
veggies_df['sunlight'] = veggies_df['sunlight'].str.strip().str.replace('\n', ' to ').str.split(' to ')

veggies_df['season'] = veggies_df['season'].str.strip().str.lower().fillna('none').str.extract(r'\((.*)\)')[0].str.split().str[0].str.split()

veggies_df['veggie'].fillna('none', inplace=True)
veggies_df['veggie'] = veggies_df['veggie'].str.strip().str.lower().str.split(' - ').str[0]

veggies_df['special characteristics'].fillna('none', inplace=True)
veggies_df['special characteristics'] = veggies_df['special characteristics'].str.strip().str.lower().str.split(' - ')

veggies_df['how to grow'].fillna('none', inplace=True)
veggies_df['how to grow'] = veggies_df['how to grow'].str.strip().str.lower().str.split('-')

veggies_df['germination temperature'] = veggies_df['germination temperature'].str.strip().str.lower().str.split(' - ').str[0].fillna('0 f to 0').apply(lambda x : ('0 f to ' + x) if len(x) == 5 else x).str.strip(' f').str.split(r'(?: f)? to ').apply(lambda x: list(map(int, x))).apply(lambda x: [0] if list(range(x[0],x[1])) == [] else list(range(x[0],x[1])))

veggies_df['tolerates'] = veggies_df['tolerates'].str.strip().str.lower().fillna('none').str.split(' - ').str[0].str.split()
veggies_df['family'] = veggies_df['family'].str.strip().str.lower().fillna('none')
veggies_df['description'] = veggies_df['description'].str.strip().str.lower().fillna('none')
veggies_df['soil conditions'] = veggies_df['soil conditions'].str.strip().str.lower().fillna('none').str.split('\n')
veggies_df['ease of care'] = veggies_df['ease of care'].str.strip().str.lower().fillna('none').replace({'easy': 'low', 'moderately difficult': 'medium', 'difficult': 'high'}).str.split()
veggies_df['foliage colors'] = veggies_df['foliage colors'].str.strip().str.lower().fillna('none').str.split().str[1].str.split()
veggies_df['lifecycle'] = veggies_df['lifecycle'].str.strip().str.lower().fillna('none').str.split()
veggies_df['family'] = veggies_df['family'].str.split(' - ').apply(lambda x : x[-1] if len(x) != 1 else x[0])
veggies_df['foliage texture'] = veggies_df['foliage texture'].str.strip().str.lower().fillna('none').str.split()

veggies_df.rename(columns={"tolerates": "problem_solvers", "ease of care": "maintenance", "lifecycle": "type", "sunlight": "sun", "veggie": "common", "soil conditions": "soil_conditions", "foliage colors": "color",
                  "special characteristics": "native_region", "how to grow": "propagation", "germination temperature": "temperature", "emergence days": "emergence_days", "season" : "bloom_time", "hardiness" : "zones", "foliage texture" : "texture"}, inplace=True)
veggies_df.drop(columns=['scene', 'url_scene',
                'seed save time', 'pests', 'hardiness_ind',"maintenance info"], inplace=True)

veggies_df = veggies_df.convert_dtypes()
```

```python
test = pd.merge(mbg_df,bhg_df,on='common', how='outer')

test['color'] = test['color_x'].combine_first(test['color_y'])
test['bloom_time'] = test['bloom_time_x'].combine_first(test['bloom_time_y'])
test['genus'] = test['genus_x'].combine_first(test['genus_y'])
test['height'] = test['height_x'].combine_first(test['height_y'])
test['width'] = test['width_x'].combine_first(test['width_y'])
test['special_feature'] = test['special_feature_x'].combine_first(test['special_feature_y'])
test['sun'] = test['sun_x'].combine_first(test['sun_y'])
test['problem_solvers'] = test['problem_solvers_x'].combine_first(test['problem_solvers_y'])
test['type'] = test['type_x'].combine_first(test['type_y'])
test['zones'] = test['zones_x'].combine_first(test['zones_y'])

test.drop(columns=['color_x','bloom_time_x','genus_x','height_x','width_x','special_feature_x','sun_x', 'problem_solvers_x', 'type_x','zones_x','color_y','bloom_time_y','genus_y','height_y','width_y','special_feature_y','sun_y', 'problem_solvers_y', 'type_y','zones_y'],inplace=True)
```

```python
whole = pd.merge(test,veggies_df,on='common',how='outer')

whole['maintenance'] = whole['maintenance_x'].combine_first(whole['maintenance_y'])
whole['native_region'] = whole['native_region_x'].combine_first(whole['native_region_y'])
whole['propagation'] = whole['propagation_x'].combine_first(whole['propagation_y'])
whole['bloom_time'] = whole['bloom_time_x'].combine_first(whole['bloom_time_y'])
whole['sun'] = whole['sun_x'].combine_first(whole['sun_y'])
whole['problem_solvers'] = whole['problem_solvers_x'].combine_first(whole['problem_solvers_y'])
whole['type'] = whole['type_x'].combine_first(whole['type_y'])
whole['zones'] = whole['zones_x'].combine_first(whole['zones_y'])
whole['family'] = whole['family_x'].combine_first(whole['family_y'])
whole['color'] = whole['color_x'].combine_first(whole['color_y'])

whole.drop(columns=['family_x','maintenance_x', 'native_region_x','propagation_x','bloom_time_x','sun_x','problem_solvers_x', 'type_x', 'zones_x', 'color_x', 'bloom_time_y','family_y','sun_y','zones_y', 'type_y', 'maintenance_y','problem_solvers_y', 'native_region_y','propagation_y','maintenance_y', 'color_y'],inplace=True)

whole['common'].fillna('none', inplace=True)
whole['family'].fillna('none', inplace=True)
whole['genus'].fillna('none', inplace=True)
whole['scientific_name'].fillna('none', inplace=True)
whole['description'].fillna('none', inplace=True)
whole['height'] = whole['height'].fillna('0').apply(lambda x: list(map(int, x)))
whole['width'] = whole['width'].fillna('0').apply(lambda x: list(map(int, x)))
whole['temperature'] = whole['temperature'].fillna('0').apply(lambda x: list(map(int, x)))
whole['emergence_days'] = whole['emergence_days'].fillna('0').apply(lambda x: list(map(int, x)))
whole['zones'] = whole['zones'].fillna('0').apply(lambda x: list(map(int, x)))
whole.fillna(whole.notna().applymap(lambda x: x or ['none']),inplace=True)

whole = whole.convert_dtypes()
```
