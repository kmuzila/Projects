```python
import numpy as np
import pandas as pd 
import sqlalchemy
import pandasql as ps

import plotly.express as px
```

```python
sqlalchemy.__version__
```

```python
sunshine = pd.read_csv('Data/sunshine_hours_by_city.csv')
sunshine.head()
```

```python
sunshine.shape
```

```python
cities = pd.read_csv('Data/cities.csv')
cities.head()
```

```python
cities.shape
```

```python
states = pd.read_csv('Data/states.csv')
states.head()
```

```python
states.shape
```

```python
countries = pd.read_csv('Data/countries.csv')
countries.head()
```

```python
countries.shape
```

```python
merge_1 = pd.merge(cities, states, left_on='state_id', right_on='id')
pd.set_option('display.max_columns', None)
merge_1.head()
```

```python
merge_2 = pd.merge(sunshine, merge_1, left_on='City', right_on='name_x')
merge_2.head()
```

```python
merge_3 = pd.merge(merge_2, countries, left_on='Country', right_on='name')
merge_3.head()
```

```python
cleaned_data = merge_3.drop(['name_x', 'country_id_x', 'country_code_x', 'country_name_x', 'wikiDataId', 'id_y', 'name_y', 'country_id_y', 'country_code_y', 'country_name_y', 'state_code_y', 'type', 'latitude_y', 'longitude_y', 'id', 'name', 'id_x', 'state_id', 'state_code_x', 'numeric_code', 'phone_code', 'currency', 'currency_name', 'currency_symbol', 'tld', 'native', 'timezones', 'latitude', 'longitude', 'emoji', 'emojiU'], axis=1)
cleaned_data.head()
```

```python
fig = px.scatter_geo(cleaned_data, lat="latitude_x", lon="longitude_x", color="Jan",
                     hover_name="City", size="Year",
                     projection="natural earth",height=700, width=1000)
fig.update_layout(
        title_text = 'World',
        showlegend = True,
        geo = dict(
            scope = 'world',
            landcolor = '#0A6007',
        )
    )
fig.update_geos(
    visible=True,
    resolution=50,
    showlakes=True, lakecolor="Blue",
    showrivers=True, rivercolor="Blue",
    showocean = True,oceancolor="#2695A0",
)
fig.show()
```

```python
fig = px.scatter_geo(cleaned_data.query("subregion=='Northern America'"), lat="latitude_x", lon="longitude_x", color="Jan",
                     hover_name="City", size="Year",
                     projection="natural earth",height=700, width=1000)
fig.update_layout(
        title_text = 'Northern America',
        showlegend = True,
        geo = dict(
            scope = 'north america',
            landcolor = '#0A6007',
        )
    )
fig.update_geos(
    visible=True,
    resolution=50,
    showlakes=True, lakecolor="Blue",
    showrivers=True, rivercolor="Blue",
    showocean = True,oceancolor="#2695A0",
)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heatmap_data = cleaned_data.filter(['Country', 'City', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'latitude_x', 'longitude_x', 'state_name'], axis=1)
# pd.set_option('display.max_rows', None)
heatmap_data_2 = heatmap_data.loc[heatmap_data['Country'] == 'United States']
heatmap_data_2['city_state'] = heatmap_data_2['City']+ ', ' + heatmap_data_2['state_name']
heatmap_data_3 = heatmap_data_2.drop(['Country', 'City', 'latitude_x', 'longitude_x', 'state_name'], axis=1).drop_duplicates(subset=['city_state'])
# heatmap_data_4 = heatmap_data_3.reset_index()
# heatmap_data_4['city_state'] = heatmap_data_4['City']+ ', ' + heatmap_data_4['state_name']
# heatmap_data_5 = heatmap_data_4.drop(['index', 'City', 'latitude_x', 'longitude_x', 'state_name'], axis=1)
# heatmap_data_6 = heatmap_data_5.set_index('city_state')
# heatmap_data_6.shape
heatmap_data_4 = heatmap_data_3.set_index('city_state')
heatmap_data_4.shape
```

```python
heatmap_data_2.head()
```

```python
fig, ax = plt.subplots(figsize=(25,90))
ax.set_title('City and Total Sun Hous per Month')
heatmap_per_month = sns.heatmap(heatmap_data_4, cbar_kws={'label': 'Total Sun Hours'}, ax=ax)
# heatmap_per_month = sns.heatmap(heatmap_data_4)
```

```python
heatmap_per_day = heatmap_data_4.div(28)
heatmap_per_day.shape
```
