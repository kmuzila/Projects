```python
import pandas as pd
import os
import json
import requests
import twint
import nest_asyncio
```

# Below are functions that can help us generate datasets

```python
def import_ac_data(label = None):
    acnh = {}
    if isinstance(label, str):
        return pd.read_csv("data\\" + label + ".csv")
    elif isinstance(label, list):
        for x in label:
            acnh[x] = pd.read_csv("data\\" + x + ".csv")
        return acnh
    else:
        for x in os.listdir('data'):
            acnh[x[:-4]] = pd.read_csv('data\\' + x)
        return acnh

def build_search(name: str = None, keywords: list = None):
    search = ''

    if name != None:
        search += name + " AND "
    if keywords != None:
        for x in range(len(keywords)):
            if x == 0:
                search += "(" + keywords[x] + " OR "
            elif x == len(keywords) - 1:
                search += keywords[x] + ")"
            else:
                search += keywords[x] + " OR "
    search += " AND (#AnimalCrossing OR #ACNH)"
    print(search)
    return search

def ascii_check(row):
    try:
        row["tweet"].encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True    

def get_tweets(name: str = None, keywords: list = None, lang: str = "en", limit: int = 20, popular: bool = False):
    search = build_search(name, keywords)
    config = twint.Config()
    config.Search = search
    config.Limit = limit
    config.Lang = lang
    config.Popular_tweets = popular
    config.Filter_retweets = True
    config.Pandas = True
    config.Hide_output = True

    nest_asyncio.apply()
    twint.run.Search(config)

    if lang == "en":
        return twint.storage.panda.Tweets_df[twint.storage.panda.Tweets_df.apply(ascii_check, axis=1)]
    else: 
        return twint.storage.panda.Tweets_df
```
